```python
import pandas as pd
df = pd.read_table(
    'BRCA_For_Algorithm.txt',header=0,sep="\t")
```

```python
df = df[:50] #用50份先做測試
```

```python
#自己寫的排序演算法 使用combsort寫
data  = df.copy()#copy 原檔案 可以讓測試檔跟原本做比較
def combsort(num):
    gap = len(num)
    swaps = True
    while gap > 1 or swaps:
        gap = max(1, int(gap / 1.25))  # minimum gap is 1
        swaps = False
        for i in range(len(num) - gap):
            j = i+gap
            if num[i] > num[j]:
                data.iloc[[i,j]] = data.reindex([j,i])
                swaps = True

combsort(data['Start'])
data
```
